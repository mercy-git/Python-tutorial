# Reading & Writing large data using pandas
larger than 1GB, lesser than 100 GB

```python
# How to read large data using pandas -> larger than 1GB, lesser than 100 GB
# reading large datasets using pandas
import pandas as pd
dfs_lst = []
# specify no of rows to read at a time in chunksize
chunks_text_reader = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv', 
                                 chunksize=10)
# each chunk_df contains 10 rows of data
for chunk_df in chunks_text_reader: 
    dfs_lst.append(chunk_df)
large_data_df = pd.concat(dfs_lst)
large_data_df
```

```python
# memory utilized by df in GB
str(large_data_df.memory_usage().sum()/1073741824) + ' GB'
```

```python
# how to write large data using pandas
large_data_df.to_csv('large_data.csv', chunksize=10, index=False)
```
